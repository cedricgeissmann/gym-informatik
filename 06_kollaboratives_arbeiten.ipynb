# Kollaboratives Arbeiten

Zusammenarbeiten ist im digitalen Zeitalter etwas vom wichtigsten das man
können muss. Es gibt sehr viele Werkzeuge die das gemeinsame arbeiten an einem
Dokument oder Projekt ermöglichen. Wenn Sie ein Word Dokument im Teams abgelegt
haben und alle dieses Dokument direkt bearbeiten, sehen Sie gerade was die
anderen machen. Das ist relativ praktisch, braucht aber eine kontinuierliche
Internetverbindung. Das bedeutet das Daten oft unnötigerweise übertragen werden
und wenn jemand über eine nicht stabile Internetverbindung verfügt, kann er an
dem Prozess nicht Teilnehmen. Wenn die Dokumente sehr gross werden und man
nicht über eine super schnelle Internetverbindung verfügt, wird das Arbeiten an
dem Dokument sehr langsam und mühsam. Das mussten jetzt sehr viele Arbeitnehmer
die ins Home-Office geschickt wurden ertragen. Diese Art der Zusammenarbeit hat
auch den weiteren Nachteil das grosse Teile der Arbeit einfach vorloren gehen
können. Wenn jemand Offline arbeitet, und nach der Bearbeitung die Datei
einfach wieder hochlädt, überschreibt er damit die Datei und allfällige
Änderungen. Dann kann man nur hoffen das Teams eine automatische Versionierung
hat. Dann steht man aber vor dem Problem das die beiden unterschiedlichen
Dateien zusammengefügt werden müssen.

Zum Glück gibt es `Git`! Denn genau für solche Workflows wurde `Git`
konzipiert. Bis jetzt wissen Sie nur das Sie mit `Git` Versionen von Ihrem
Projekt erstellen können, aber zusammenarbeit mit `Git` haben wir uns noch
nicht angeschaut.

## Git

`Git` ist ein sogenannt dezentrales Versionierungssystem. Das löst genau dieses
Problem das oben beschrieben wurde. Mit `Git` kann jeder der an einem Projekt
mitarbeiten möchte, das `remote` Repository `klonen` (sich auf den eigenen
Rechner kopieren) und daran arbeiten. Wenn man mit seiner Arbeit fertig ist
sendet man den aktuellen Stand des Projekts wieder auf des `remote` Repository,
wo auch andere daran arbeiten können. Sie brauchen also während dem arbeiten
keine aktive Internetverbindung.

Damit sind wir aber genau in dem problematischen Fall von oben, wo jemand eine
neue Datei hochlädt und alles überschreibt. Das kann man leider nicht umgehen,
aber `Git` bietet sehr gute Werkzeuge an um damit umzugehen. In diesem Szenario
können 3 Fälle auftauchen. Wir werden hier alle besprechen, und Sie werden dann
jeweils eine Aufgabe dazu machen.

### Fall 1; keine Änderungen

Das ist der einfachste Fall. Hier gab es auf dem `remote` keine Änderungen.
`Git` erkennt das und lädt Ihren aktuellen stand einfach hoch. Sie müssen
nichts machen, `Git` macht die ganze Arbeit von alleine.

### Fall 2; Änderungen, aber in anderen Teilen

Wenn `Git` erkennt das es neuere Versionen auf dem `remote` gibt wie Sie lokal
bei sich haben, dann müssen Sie diese zuerst herunterladen (dazu sagt man
`pullen`). `Git` verbietet Ihnen in diesem Fall auf das `remote` hochzuladen
(dazu sagt man `pushen`), bis Sie die Änderungen `gepullt` haben. Das tolle
daran ist das `Git` dann von alleine erkennt das die Änderungen an
verschiedenen Stellen vorkommen und es kann automatisch einen sogenannten
`merge` machen. Danach können Sie einfach `pushen` und alles ist in bester
Ordnung, und es wurden alle Änderungen von allen berücksichtigt. Also auch in
diesem Fall macht `Git` die ganze Arbeit für Sie.

### Fall 3; `merge` Konflikt

In diesem Fall ist `Git` leider nicht imstande alles für Sie zu übernehmen.
`Git` hat erkannt das jemand anderes eine Version `gepusht` hat, die Änderungen
an den gleichen Stellen wie Sie vorgenommen hat. Das nennt man einen `merge
Konflikt` und der ist bei den Programmierern nicht so gerne gesehen, denn es
bedeutet Arbeit für sie, weil jemand anderes schneller war... Meist ist es aber
nicht so schlimm, und `merge Konflikte` können sehr schnell behoben werden.

Wenn Sie einen `merge Konflikt` haben, wird `Git` Ihnen sagen an welcher Stelle
sich der Konflikt befindet und zeigt Ihnen gerade die beiden Änderungen an. Sie
müssen dann entscheiden welche Änderung Sie am Ende im Projekt haben wollen,
oder ob es sogar eine Mischung aus beiden Änderungen sein soll. Dazu verändern
Sie die Datei einfach so dass am Ende das drin steht was Sie möchten, und
machen dann einen neuen `commit` wenn alle Ihre Änderungen gespeichert sind.
Dann können Sie `pushen` und alles ist wie es sein soll.

## Fälle üben

Den Fall 1 üben wir nicht mehr, denn das haben Sie beim Blatt 5 schon
ausführlich gemacht. Wir üben aber die Fälle 2 und 3.

**Aufgabe:**

> In der Datei `README.md` machen Sie 2 Änderungen. Die erste Änderung machen
> Sie am Titel und zwar müssen Sie das auf `Github` machen. Damit simulieren
> Sie eine Änderung die von jemand anderem kommt.
>
> Die zweite Änderung machen Sie lokal in `VSCodium` und zwar an einer
> **anderen** Stelle wie im Titel. Dann versuchen Sie zu `pushen`.

Wenn Sie alles richtig gemacht haben, sollten Sie von `Git` aufgefordert worden
sein die Änderungen zuerst herunter zu laden, und `Git` konnte dann einen
`merge` machen. Dann sollten Sie `pushen` können.

**Aufgabe:**

> Nun wollen wir einen `merge Konflikt` erzeugen. Dazu gehen wir fast gleich
> vor wie in der letzten Aufgabe, aber diesmal machen wir beide Änderungen im
> Titel.
>
> Ändern Sie auf `Github` den Titel zu `Titel von Github`. Und in `VSCodium`
> ändern Sie den Titel zu `Titel von VSCodium`. Machen Sie auch davon einen
> neuen `commit` und versuchen Sie zu `pushen`.
>
> Wenn Sie beim `merge Konflikt` angelangt sind, diskutieren Sie mit Ihrem
> Sitznachbarn was genau Sie sehen, und versuchen Sie daraus Sinn zu machen.
>
> Lösen Sie den `merge Konflikt` und `pushen` Sie Ihre Änderung.

## Zusammenarbeiten

Sie können auch an Repositories arbeiten die nicht Ihre eigenen sind. Wie gut
das jeweils funktioniert hängt davon ab wie Sie die Zugriffsrechte für das
Repository eingestellt haben.

**Aufgabe:**

> Bilden Sie 2er Gruppen. In der Gruppe machen Sie ab wer ein neues Repository
> erstellt. Diese Person erstellt das neue Repository an dem beide arbeiten
> werden.

## Repository `klonen`

Der normale Workflow ist ein wenig anders wie wir das beim letzten mal gemacht
haben. Normalerweise starten Sie mit einem Repository das Sie vom `remote`
`klonen` und nicht mit einem das Sie lokal erstellt haben. Das geht mit
`VSCodium` ganz einfach. Öffnen Sie ein neues Fenster (`File > New Window`).
Dann gehen Sie auf die Repository Ansicht, und dort sollten Sie die
Schaltfläche `Clone Repository` finden. Klicken Sie diese an und `VSCodium`
wird Sie nach der URL fragen. Die finden Sie auf `Github` bei dem neuen
Repository das Sie erstellt haben.

**ACHTUNG!**
> Auch wie schon bei Blatt 5, fügen Sie hier Ihren `Guthub`-Benutzernamen ein,
> dass sollte dann in etwa so aussehen:

```bash
https://benutzername@github.com/benutzername/reponame.git
```

**Aufgabe:**

> Jemand von Ihnen setzt das Repository auf. Das heisst **nur einer** erstellt
> die Datei `README.md` und fügt einen Titel ein. Dann machen Sie den ersten
> `commit` und `pushen` auf das Repository.
>
> Jetzt kann die Andere Person `pullen` und bekommt die Änderungen gleich mit.

## Zusammenfassung in Markdown

Wie Sie im letzten Kapitel bereits gesehen haben, können Sie Markdown verwenden
um Dokumente zu schreiben. Das praktische daran ist, dass Sie sich nicht
dauernd um die Gestaltung kümmern müssen. Sobald Sie die Syntax mal ein wenig
verstanden haben, können Sie Dokumente sehr schnell und einfach schreiben. Es
gibt dann Programme die Ihnen richtige Dokumente aus Ihrem Markdown erstellen
(`pandoc` wäre ein tolles Beispiel). Wir machen es uns aber relativ einfach,
und nutzen nur das Feature von `Github` dass die `README.md` Datei automatisch
als Dokument dargestellt wird.

**Aufgabe:**

> Sie sollen nun in Ihrem 2er Team eine Zusammenfassung von all dem machen was
> wir hier bereits gelernt haben. Dazu sollen Sie die Arbeit ein wenig
> aufteilen, es müssen also nicht beide zu allem etwas schreiben. Koordinieren
> Sie das zusammen.
> 
> Sie sollen nach jedem Abschnitt den Sie geschrieben haben, einen neuen
> `commit` machen. Sie müssen nicht jedesmal `puschen` denn Sie werden so recht
> viele `merge Konflikte` erhalten. `Pushen` Sie ca. all 3 Abschnitte und
> koordinieren Sie sich auch hier. Denn es wir immer der langsamere sein der
> die `merge Konflikte` bekommt.
>
> Die Strategie für `merge Konflikte` ist immer beides zu behalten, ausser es
> wurde über das gleiche Thema geschrieben (wenn die Titel ähnlich sind). Dann
> müssen Sie auswählen welche Teile von den verschiedenen VErsionen sie haben
> möchten.
